```python
import pandas
import numpy as np
import matplotlib.pyplot as plot
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
import os


print(os.path)
```

```python
data = pandas.read_csv('co2_emissions_data.csv')

# Step1 - Data analysis
# print(data.head())

trainingData, testData = train_test_split(data, test_size=0.2, random_state=100)

print(trainingData.shape)
print(testData.shape)

```

```python
def normalize(xMin, xMax, x):
    return (x - xMin) / (xMax - xMin)

def denormalize(xMin, xMax, x):
    return x * (xMax - xMin) + xMin

   
```

```python
# check for missing values in the data
# print(data.isnull().sum())


# data.describe()

co2 = trainingData['CO2 Emissions(g/km)']
# print(co2.describe())

minCo2 = co2.min()
maxCo2 = co2.max()

print(co2.size)
print(co2[0])

for i in range(len(co2)):
    co2[i] = normalize(minCo2, maxCo2, co2.iloc[i])
print(co2.describe())

trainingData.hist(figsize=(10, 8))
plot.show()    
    
```
